```python
import pandas as pd

# Read the uploaded Excel file
df = pd.read_excel('/content/Financial_Data_Tesla_Microsoft_Apple.xlsx')

# Display the first few rows
df.head()
```

```python
# Make sure Fiscal Year is an integer
df['Fiscal Year'] = df['Fiscal Year'].astype(int)

# Sort by Company and Year
df = df.sort_values(by=['Company', 'Fiscal Year'])

# Reset index for clarity
df = df.reset_index(drop=True)

# Display the sorted table
df
```

```python
# Calculate YoY Growth %
df['Revenue Growth (%)'] = df.groupby('Company')["Total Revenue (USD Millions)"].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby('Company')["Net Income (USD Millions)"].pct_change() * 100
df['Assets Growth (%)'] = df.groupby('Company')["Total Assets (USD Millions)"].pct_change() * 100
df['Liabilities Growth (%)'] = df.groupby('Company')["Total Liabilities (USD Millions)"].pct_change() * 100
df['Cash Flow Growth (%)'] = df.groupby('Company')["Cash Flow from Ops (USD Millions)"].pct_change() * 100

# Display the updated DataFrame
df
```

# Financial Analysis of Microsoft, Tesla, and Apple (Last 3 Fiscal Years)

## Data Source
- Extracted key financial data from SEC 10-K filings for Microsoft, Tesla, and Apple.
- Metrics analyzed: Total Revenue, Net Income, Total Assets, Total Liabilities, Cash Flow from Operating Activities.

## Data Preparation
- Data loaded from an Excel file.
- Sorted by Company and Fiscal Year for consistency.
- Calculated year-over-year growth percentages for each financial metric.

## Key Observations
- Tesla showed high growth in revenue and net income, especially from 2021 to 2022.
- Microsoft maintained steady growth with strong cash flow from operations.
- Apple’s revenue fluctuated slightly, but net income remained strong.

```python
def simple_chatbot():
    print("Welcome to the Financial Chatbot! Ask me about Microsoft, Tesla, or Apple.")
    print("Supported queries include:")
    print("1. What is the total revenue?")
    print("2. How has net income changed over the last year?")
    print("3. What is the cash flow from operating activities?")
    print("4. Which company had the highest revenue last year?")
    print("5. What is the total assets for Microsoft in 2023?")
    print("Type 'exit' to quit.\n")

    while True:
        user_query = input("You: ").lower().strip()

        if user_query == 'exit':
            print("Chatbot: Goodbye!")
            break

        elif "total revenue" in user_query:
            latest_year = df['Fiscal Year'].max()
            response = f"Total Revenue in {latest_year}:\n"
            for company in df['Company'].unique():
                revenue = df[(df['Company'] == company) & (df['Fiscal Year'] == latest_year)]['Total Revenue (USD Millions)'].values[0]
                response += f"- {company}: ${revenue:,} million\n"
            print("Chatbot:", response)

        elif "net income changed" in user_query or "net income" in user_query:
            latest_year = df['Fiscal Year'].max()
            response = f"Net Income change from {latest_year - 1} to {latest_year}:\n"
            for company in df['Company'].unique():
                growth = df[(df['Company'] == company) & (df['Fiscal Year'] == latest_year)]['Net Income Growth (%)'].values[0]
                growth_text = f"{growth:.2f}%" if not pd.isna(growth) else "No data"
                response += f"- {company}: {growth_text}\n"
            print("Chatbot:", response)

        elif "cash flow" in user_query:
            latest_year = df['Fiscal Year'].max()
            response = f"Cash Flow from Operating Activities in {latest_year}:\n"
            for company in df['Company'].unique():
                cf = df[(df['Company'] == company) & (df['Fiscal Year'] == latest_year)]['Cash Flow from Ops (USD Millions)'].values[0]
                response += f"- {company}: ${cf:,} million\n"
            print("Chatbot:", response)

        elif "highest revenue" in user_query:
            latest_year = df['Fiscal Year'].max()
            subset = df[df['Fiscal Year'] == latest_year]
            highest = subset.loc[subset['Total Revenue (USD Millions)'].idxmax()]
            response = f"The company with the highest revenue in {latest_year} is {highest['Company']} with ${highest['Total Revenue (USD Millions)']:,} million."
            print("Chatbot:", response)

        elif "total assets for microsoft in 2023" in user_query:
            assets = df[(df['Company'] == "Microsoft") & (df['Fiscal Year'] == 2023)]['Total Assets (USD Millions)'].values[0]
            response = f"Microsoft's Total Assets in 2023 are ${assets:,} million."
            print("Chatbot:", response)

        else:
            print("Chatbot: Sorry, I can only answer predefined financial queries.")
```

```python
simple_chatbot()
```

## Supported Queries:
- Total revenue for the latest fiscal year
- Year-over-year net income changes
- Cash flow from operating activities
- Highest revenue company last year
- Microsoft’s total assets in 2023

## Limitations:
